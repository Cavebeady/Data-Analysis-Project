* Importing libraries

```python
import numpy as np
import pandas as pd
from datetime import datetime, timedelta
import os
```

+ Creating seed

```python
np.random.seed(42)
N_TRANSACTIONS = 5000
os.makedirs('data', exist_ok = True)
os.makedirs('outputs/reports', exist_ok = True)
```

* Generating Product Catalog

```python
product_names = ['Laptop', 'Wireless Mouse', 'USB-C Hub', 'Monitor', 'Keyboard', 'Webcam', 'Headphones', 'Phone Stand', 'Desk Lamp', 'Notebook']
categories = ['Electronics', 'Electronics', 'Electronics', 'Electronics', 'Electronics', 'Electronics', 'Audio', 'Accessories', 'Accessories', 'Stationery']
base_prices = np.array([999.99, 29.99, 49.99, 349.99, 79.99, 89.99, 149.99, 19.99, 39.99, 9.99])

products_df = pd.DataFrame({'product_id': [f'P{i:03d}' for i in range(1, 11)], 'product_name': product_names, 'category': categories, 'base_price': base_prices})
products_df.to_csv('data/products.csv', index = False)
print('Products saved:', products_df.shape)
```

* Generating customer profiles

```python
N_CUSTOMERS = 500
cities = ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix']
tiers = ['Bronze', 'Silver', 'Gold', 'Platinum']
customers_df = pd.DataFrame({
 'customer_id': [f'C{i:04d}' for i in range(1, N_CUSTOMERS + 1)],
 'city': np.random.choice(cities, N_CUSTOMERS),
 'tier': np.random.choice(tiers, N_CUSTOMERS, p=[0.4, 0.3, 0.2, 
0.1]),
 'age': np.random.randint(18, 75, N_CUSTOMERS),
 'join_year': np.random.randint(2018, 2024, N_CUSTOMERS)
})
customers_df.to_csv('data/customers.csv', index=False)
print('Customers saved:', customers_df.shape)
```

* Generating Sales Transactions

```python
start_date = datetime(2023, 1, 1)
date_offsets = np.random.randint(0, 365, N_TRANSACTIONS)
dates = [start_date + timedelta(days=int(d)) for d in date_offsets]
product_ids = np.random.choice(products_df['product_id'], 
N_TRANSACTIONS)
customer_ids = np.random.choice(customers_df['customer_id'], 
N_TRANSACTIONS)

quantities = np.random.choice(
 [1, 2, 3, 4, 5], N_TRANSACTIONS, p=[0.5, 0.25, 0.12, 0.08, 0.05]
)
discount_pct = np.random.choice(
 [0, 5, 10, 15, 20], N_TRANSACTIONS, p=[0.5, 0.2, 0.15, 0.1, 0.05]
)
sales_df = pd.DataFrame({
 'transaction_id': [f'T{i:05d}' for i in range(1, N_TRANSACTIONS + 
1)],
 'date': dates,
 'product_id': product_ids,
 'customer_id': customer_ids,
 'quantity': quantities,
 'discount_pct': discount_pct
})
# Injecting ~2% nulls for cleaning practice
null_idx = np.random.choice(sales_df.index, size=100, replace=False)
sales_df.loc[null_idx[:50], 'quantity'] = np.nan
sales_df.loc[null_idx[50:], 'discount_pct'] = np.nan
sales_df.to_csv('data/sales_data.csv', index=False)
print('Sales saved:', sales_df.shape)
```

# Data Cleaning and Preprocessing
* Loading and Inspecting

```python
import numpy as np
import pandas as pd

sales = pd.read_csv('data/sales_data.csv')
products = pd.read_csv('data/products.csv')
customers = pd.read_csv('data/customers.csv')
print('Shape:', sales.shape)
print(sales.dtypes)
print('Nulls:\n', sales.isnull().sum())
```

+ Fixing date type and engineer features

```python
sales['date'] = pd.to_datetime(sales['date'])
sales['month'] = sales['date'].dt.month
sales['month_name'] = sales['date'].dt.strftime('%b') # .strftime() stands for 'string format time'
sales['quarter'] = sales['date'].dt.quarter
sales['day_of_week'] = sales['date'].dt.day_name()
sales['week_number'] = sales['date'].dt.isocalendar().week.astype(int)
```

+ Handling missing values

```python
# Filling quantity with median 
median_qty = sales['quantity'].median()
sales['quantity'] = sales['quantity'].fillna(median_qty).astype(int)
# Filling discount_pct with 0 (assumes no discount if unknown)
sales['discount_pct'] = sales['discount_pct'].fillna(0)
print('Nulls after fill:', sales.isnull().sum().sum())
```

* Merging all three datasets


```python
# Left join: keep all sales rows
df = sales.merge(products, on='product_id', how='left')
df = df.merge(customers, on='customer_id', how='left')
print('Master shape:', df.shape)
print('Columns:', df.columns.tolist())
```

* Computing revenue

```python
# Revenue = quantity x base_price x (1 - discount/100)
df['revenue'] = (
 df['quantity'] * df['base_price'] *
 (1 - df['discount_pct'] / 100)
).round(2)
```

* Detecting outliers with NumPy IQR

```python
revenue_arr = df['revenue'].values
Q1 = np.percentile(revenue_arr, 25)
Q3 = np.percentile(revenue_arr, 75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outlier_mask = (df['revenue'] < lower_bound) | (df['revenue'] > 
upper_bound)
print(f'Outliers: {outlier_mask.sum()} rows ({outlier_mask.mean()*100:.1f}%)')
# Cap outliers instead of dropping
df['revenue_capped'] = df['revenue'].clip(lower_bound, upper_bound)
```

* Validating and saving

```python
assert df.isnull().sum().sum() == 0, 'Nulls remain!'
assert (df['revenue'] >= 0).all(), 'Negative revenue!'
assert not df.duplicated('transaction_id').any(), 'Duplicate IDs!'
print('All validation checks passed.')
df.to_csv('data/master_data.csv', index=False)
```

fillna() fills any null values in a row/column with the selected value   while dropna() removes any row/column that has a null value. dropna is used when the user is okay with deleting entire rows from their data

Median isn't easily affected by ouliers unlike the mean value, so it's used when filling NaN values

left merge includes everything in the left data frame while also including what is common between both data frames.  The number of rows in the result was 5000

# NumPy Deep Dive — Statistical Analysis
* Loading data as NumPy arrays

```python
df = pd.read_csv('data/master_data.csv', parse_dates=['date'])
# Extracting NumPy arrays
revenue = df['revenue'].values
quantity = df['quantity'].values
prices = df['base_price'].values
discounts = df['discount_pct'].values
months = df['month'].values
```

* Full Descriptive Statistics

```python
def describe_array(arr, label):
 print(f'\n=== {label} ===')
 print(f' Count : {len(arr):>10,}')
 print(f' Mean : {np.mean(arr):>10.2f}')
 print(f' Median : {np.median(arr):>10.2f}')
 print(f' Std Dev : {np.std(arr):>10.2f}')
 print(f' Min : {np.min(arr):>10.2f}')
 print(f' Max : {np.max(arr):>10.2f}')
 print(f' Range : {np.ptp(arr):>10.2f}')
 print(f' P25 : {np.percentile(arr, 25):>10.2f}')
 print(f' P75 : {np.percentile(arr, 75):>10.2f}')
 print(f' P90 : {np.percentile(arr, 90):>10.2f}')
describe_array(revenue, 'REVENUE ($)')
describe_array(quantity, 'QUANTITY (units)')
describe_array(discounts, 'DISCOUNT (%)')
```

* Monthly Revenue Aggregation

```python
monthly_revenue = np.array([
 revenue[months == m].sum() for m in range(1, 13)
])
month_names = ['Jan','Feb','Mar','Apr','May','Jun',
 'Jul','Aug','Sep','Oct','Nov','Dec']
print('\n=== MONTHLY REVENUE ===')
for i, (name, rev) in enumerate(zip(month_names, monthly_revenue), 1):
 bar = '#' * int(rev / monthly_revenue.max() * 30)
 print(f' {name}: ${rev:>10,.2f} {bar}')
best_month_idx = np.argmax(monthly_revenue)
worst_month_idx = np.argmin(monthly_revenue)

print(f'\nBest month: {month_names[best_month_idx]} (${monthly_revenue[best_month_idx]:,.2f})')
print(f'Worst month: {month_names[worst_month_idx]} (${monthly_revenue[worst_month_idx]:,.2f})')
```

* 3-Month Rolling Average

```python
window = 3
weights = np.ones(window) / window
rolling_avg = np.convolve(monthly_revenue, weights, mode='valid')
print('\n=== 3-MONTH ROLLING AVERAGE ===')
for i, avg in enumerate(rolling_avg):
 print(f' {month_names[i+1]}: ${avg:>10,.2f}')
```

* Broadcasting: Tier Multiplier

```python
tier_map = {'Bronze': 1.0, 'Silver': 1.05, 'Gold': 1.10, 'Platinum': 1.15}
multiplier = df['tier'].map(tier_map).values
# NumPy broadcasting — no Python loops
adjusted_revenue = revenue * multiplier
print(f'\nOriginal total revenue : ${revenue.sum():>12,.2f}')
print(f'Tier-adjusted revenue : ${adjusted_revenue.sum():>12,.2f}')
print(f'Tier uplift : ${adjusted_revenue.sum()-revenue.sum():>12,.2f}')
```

* Correlation Matrix

```python
corr = np.corrcoef([prices, quantity, discounts])
labels = ['Price', 'Quantity', 'Discount']
print('\n=== PEARSON CORRELATION MATRIX ===')
header = f' {'':12s}' + ''.join(f'{l:>12s}' for l in labels)
print(header)
for i, row in enumerate(corr):
 print(f' {labels[i]:12s}' + ''.join(f'{v:>12.3f}' for v in row))
```

* Top 10 Highest Revenue Transactions

```python
top10_idx = np.argsort(revenue)[-10:][::-1]
print('\n=== TOP 10 REVENUE TRANSACTIONS ===')
for rank, idx in enumerate(top10_idx, 1):
 tid = df.iloc[idx]['transaction_id']
 rev = revenue[idx]
 prod = df.iloc[idx]['product_name']
 print(f' {rank:>2}. {tid} {prod:20s} ${rev:>8,.2f}')
```

* Flag high-value transactions with np.where

```python
threshold = np.percentile(revenue, 90)
df['value_segment'] = np.where(revenue >= threshold, 'High-Value', 
'Standard')
high_val_count = (df['value_segment'] == 'High-Value').sum()
high_val_rev = revenue[revenue >= threshold].sum()
print(f'\nHigh-value transactions (top 10%): {high_val_count}')
print(f'Revenue from high-value transactions: ${high_val_rev:,.2f}')
```

np.mean() calculates average of the column i.e sum of the values divided by the number of values    
np.median() returns the middle value gotten when the column is sorted from the lowest to highest value    
Median is more appropriate because it isn't easily affected by outliers compared to the mean

np.convolve() rolls the window over the array and calculates the product of the sum of the values
 
 by using mode = valid, the window is applied to only the parts of the array it can fit in, its used when you want to avoid getting data
for periods that don't exist

there is no meaningful correlation between them  
 the value being 0.008 suggests that they had little to no impact on each other   
 
this is due to the fact a random number generator was used in making the data

# Pandas Deep Dive — Business Insights
* Revenue and Volume by Category

```python
import pandas as pd
import numpy as np
import os
df = pd.read_csv('data/master_data.csv', parse_dates=['date'])
os.makedirs('outputs', exist_ok=True)

category_summary = (
 df.groupby('category')
 .agg(
 total_revenue = ('revenue', 'sum'),
 avg_order_value = ('revenue', 'mean'),
 total_units = ('quantity', 'sum'),
 num_transactions = ('transaction_id', 'count'),
 unique_customers = ('customer_id', 'nunique'),
 avg_discount = ('discount_pct', 'mean')
 )
 .sort_values('total_revenue', ascending=False)
 .round(2)
)
# Add revenue share column
category_summary['revenue_share_pct'] = (
 category_summary['total_revenue'] /
 category_summary['total_revenue'].sum() * 100
).round(1)
print(category_summary.to_string())
category_summary.to_csv('outputs/category_summary.csv')
```

* Top 10 Customers by Lifetime Value

```python
customer_ltv = (
 df.groupby(['customer_id', 'city', 'tier'])
 .agg(
 lifetime_value = ('revenue', 'sum'),
 total_orders = ('transaction_id', 'count'),
 avg_order_value = ('revenue', 'mean'),
 total_units = ('quantity', 'sum'),
 first_purchase = ('date', 'min'),
 last_purchase = ('date', 'max'),
 )
 .reset_index()
 .sort_values('lifetime_value', ascending=False)
 .head(10)
 .round(2)
)
print(customer_ltv[['customer_id','city','tier',
 
'lifetime_value','total_orders']].to_string(index=False))
customer_ltv.to_csv('outputs/top10_customers.csv', index=False)
```

* Month-over-Month Revenue Trend 

```python
monthly = (df.set_index('date').resample('ME').agg(monthly_revenue = ('revenue', 'sum'),num_transactions = ('transaction_id', 'count'),
 avg_order_value = ('revenue', 'mean'),
 total_units = ('quantity', 'sum')))
monthly['mom_change_pct'] = monthly['monthly_revenue'].pct_change() * 100
monthly['cum_revenue'] = monthly['monthly_revenue'].cumsum()
monthly['rolling_3m_avg'] = monthly['monthly_revenue'].rolling(3).mean()
monthly = monthly.round(2)
print(monthly.to_string())
monthly.to_csv('outputs/monthly_trend.csv')
```

* Tier vs Average Order Value (Pivot Table)

```python
tier_pivot = pd.pivot_table(
    df,
    values='revenue',
    index='tier',
    columns='quarter',
    aggfunc=['mean', 'sum', 'count'], 
    margins=True,
    margins_name='All Tiers'
).round(2)

print(tier_pivot.to_string())
tier_pivot.to_csv('outputs/tier_pivot.csv')
```

* Best Product per City

```python
city_product_rev = (
 df.groupby(['city', 'product_name'])['revenue']
 .sum()
 .unstack(fill_value=0)
)
best_product_per_city = pd.DataFrame({
 'best_product' : city_product_rev.idxmax(axis=1),
 'top_revenue' : city_product_rev.max(axis=1).round(2),
 'num_products_sold': (city_product_rev > 0).sum(axis=1)
})
print(best_product_per_city.to_string())
best_product_per_city.to_csv('outputs/best_product_per_city.csv')
```

* Discount Level Impact on Revenue

```python
# Bin discount levels
df['discount_band'] = pd.cut(
 df['discount_pct'],
 bins=[-1, 0, 5, 10, 15, 20],
 labels=['No Discount', '1-5%', '6-10%', '11-15%', '16-20%']
)
discount_impact = (
 df.groupby('discount_band', observed=True)
 .agg(
 num_transactions = ('transaction_id', 'count'),
 total_revenue = ('revenue', 'sum'),
 avg_revenue = ('revenue', 'mean'),
 avg_quantity = ('quantity', 'mean'),
 )
 .round(2)
)
print(discount_impact.to_string())
discount_impact.to_csv('outputs/discount_impact.csv')
```

* Product Performance Matrix

```python
product_matrix = (
 df.groupby(['product_name', 'category'])
 .agg(
 units_sold = ('quantity', 'sum'),
 total_revenue = ('revenue', 'sum'),
 avg_discount = ('discount_pct', 'mean'),
 unique_customers = ('customer_id', 'nunique'),
 num_transactions = ('transaction_id', 'count'),
 )
 .sort_values('total_revenue', ascending=False)
 .round(2)
)
product_matrix['revenue_per_unit'] = (
 product_matrix['total_revenue'] / product_matrix['units_sold']
).round(2)
print(product_matrix.to_string())
product_matrix.to_csv('outputs/product_matrix.csv')
```

pct_change() is what calculated the month over month revenue, used in checking the growth over a month. 

it sums the current and previous revenue, divides the sum by the previous and multiplies it by 100 percent. 

The months with negative growth were february, march, may, june, august and october

# Report Generation & Excel Export
+ Setup and helpers

```python
import pandas as pd
import numpy as np
from tabulate import tabulate
from openpyxl import load_workbook
from openpyxl.styles import Font, PatternFill, Alignment, Border, Side
from openpyxl.utils import get_column_letter
import os
import datetime
# Load master data and all pre-computed outputs
df = pd.read_csv('data/master_data.csv', 
parse_dates=['date'])
category_sum = pd.read_csv('outputs/category_summary.csv', 
index_col=0)
top_customers = pd.read_csv('outputs/top10_customers.csv')
monthly = pd.read_csv('outputs/monthly_trend.csv', index_col=0)
product_matrix = pd.read_csv('outputs/product_matrix.csv')
discount_impact = pd.read_csv('outputs/discount_impact.csv')
best_per_city = pd.read_csv('outputs/best_product_per_city.csv', 
index_col=0)
os.makedirs('outputs/reports', exist_ok=True)
```

* Helper: ASCII section divider

```python
def section(title, width=60):
 line = '=' * width
 return f'\n{line}\n {title}\n{line}'
```

* Executive Summary

```python
total_revenue = df['revenue'].sum()
total_transactions = len(df)
avg_order_value = df['revenue'].mean()
total_units = df['quantity'].sum()
unique_customers = df['customer_id'].nunique()
best_category = category_sum['total_revenue'].idxmax()
best_month_idx = monthly['monthly_revenue'].idxmax()
best_customer_id = top_customers.iloc[0]['customer_id']
best_customer_ltv = top_customers.iloc[0]['lifetime_value']
# Month-over-month changes
mom = monthly['mom_change_pct'].dropna()
positive_months = (mom > 0).sum()
negative_months = (mom < 0).sum()
best_growth_month = mom.idxmax()
worst_growth_month = mom.idxmin()
report_date = datetime.date.today().strftime('%B %d, %Y')
exec_summary = f'''SHOPSMART INC. — ANNUAL SALES REPORT 2023
Report Generated: {report_date}
Prepared by: Data Analytics Team
{section('FINANCIAL HIGHLIGHTS')}
 Total Revenue : ${total_revenue:>12,.2f}
 Total Transactions : {total_transactions:>12,}
 Total Units Sold : {total_units:>12,}
 Average Order Value : ${avg_order_value:>12,.2f}
 Unique Customers : {unique_customers:>12,}
{section('TOP PERFORMERS')}
 Best Category : {best_category}
 Best Month : {best_month_idx}
 Top Customer : {best_customer_id} (LTV: 
${best_customer_ltv:,.2f})
{section('MONTH-OVER-MONTH GROWTH')}
 Months with growth : {positive_months} of 11
 Months with decline : {negative_months} of 11
 Strongest growth month: {best_growth_month}
 Weakest growth month : {worst_growth_month}
{section('REVENUE BY CATEGORY')}
'''
# Append category table
cat_display = category_sum[['total_revenue','revenue_share_pct','num_transactions']].copy()
cat_display.columns = ['Total Revenue ($)', 'Share (%)', 
'Transactions']
exec_summary += tabulate(cat_display, headers='keys', tablefmt='grid',
 floatfmt='.2f') + '\n'
# Save report
with open('outputs/reports/executive_summary.txt', 'w') as f:
 f.write(exec_summary)
print('Executive summary saved.')
```

* Category Detailed Report

```python
cat_report = section('SHOPSMART INC. — CATEGORY PERFORMANCE REPORT 2023')
cat_report += '\n\n1. FULL CATEGORY BREAKDOWN\n'
cat_report += tabulate(
 category_sum.reset_index(),
 headers=['Category','Revenue','Avg Order','Units','Transactions',
 'Customers','Avg Discount','Share %'],
 tablefmt='grid', floatfmt='.2f'
) + '\n'
# Product-level breakdown
cat_report += '\n2. PRODUCT PERFORMANCE WITHIN CATEGORIES\n'
for cat, grp in product_matrix.groupby('category'):
 cat_report += f'\n [{cat}]\n'
 sub = grp[['product_name','units_sold','total_revenue',
 'revenue_per_unit','unique_customers']].copy()
 cat_report += tabulate(sub, headers='keys', tablefmt='simple',
 showindex=False, floatfmt='.2f') + '\n'
with open('outputs/reports/category_report.txt', 'w') as f:
 f.write(cat_report)
print('Category report saved.')
```

+ Applying Excel formatting with openpyxl

```python
wb = load_workbook(excel_path)
header_fill = PatternFill('solid', fgColor='2B5797')
header_font = Font(bold=True, color='FFFFFF', size=11)
alt_fill = PatternFill('solid', fgColor='EEF3FB')
thin = Side(border_style='thin', color='CCCCCC')
cell_border = Border(left=thin, right=thin, top=thin, bottom=thin)
for sheet_name in wb.sheetnames:
 ws = wb[sheet_name]
 # Style header row
 for cell in ws[1]:
  cell.fill = header_fill
  cell.font = header_font
  cell.alignment = Alignment(horizontal='center', 
  vertical='center')
  cell.border = cell_border
 # Alternate row shading and borders for data rows
 for row_idx, row in enumerate(ws.iter_rows(min_row=2), start=2):
  fill = alt_fill if row_idx % 2 == 0 else PatternFill()
  for cell in row:
    cell.fill = fill
    cell.border = cell_border
 # Auto-size columns
 for col in ws.columns:
  max_len = max(
  len(str(c.value)) if c.value else 0 for c in col)
  ws.column_dimensions[get_column_letter(col[0].column)].width = min(max_len + 4, 40)
wb.save(excel_path)
print(f'Excel workbook saved: {excel_path}')
```
